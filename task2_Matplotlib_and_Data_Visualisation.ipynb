```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample data for demonstration
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 2, 4, 8, 16]

# Set axes label and limits
plt.figure()
plt.plot(x, y1, label='y1')
plt.plot(x, y2, label='y2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xlim(0, 6)
plt.ylim(0, 30)
plt.legend()

# Create a figure with multiple plots using subplot
plt.figure()
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('Plot 1')

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Plot 2')

plt.subplot(2, 2, 3)
plt.scatter(x, y1)
plt.title('Scatter Plot')

plt.subplot(2, 2, 4)
plt.bar(x, y1)
plt.title('Bar Plot')

# Add a legend to the plot
plt.figure()
plt.plot(x, y1, label='y1')
plt.plot(x, y2, label='y2')
plt.legend()

# Save your plot as png
plt.savefig('plot.png')

# Load iris dataset for scatter plot
iris = sns.load_dataset('iris')

# Scatter plot
plt.figure()
sns.scatterplot(x='sepal_length', y='sepal_width', data=iris, hue='species', style='species', size='petal_length')
plt.title('Scatter Plot with Iris Dataset')

# Bar plot - Simple
plt.figure()
sns.barplot(x='species', y='petal_length', data=iris)
plt.title('Simple Bar Plot')

# Bar plot - Grouped
plt.figure()
sns.barplot(x='species', y='petal_length', hue='species', data=iris)
plt.title('Grouped Bar Plot')

# Bar plot - Stacked
plt.figure()
sns.barplot(x='species', y='petal_length', data=iris, estimator=sum, errorbar=None)
plt.title('Stacked Bar Plot')

# Histogram
plt.figure()
sns.histplot(iris['petal_length'], bins=20, kde=True)
plt.title('Histogram')

# Pie plot
plt.figure()
iris['species'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Pie Plot')

# Box plot
plt.figure()
sns.boxplot(x='species', y='petal_length', data=iris)
plt.title('Box Plot')

# Violin plot
plt.figure()
sns.violinplot(x='species', y='petal_length', data=iris)
plt.title('Violin Plot')

# Contour plot
plt.figure()
sns.kdeplot(x=iris['sepal_length'], y=iris['sepal_width'], fill=True)
plt.title('Contour Plot')

# Heatmap
plt.figure()
numeric_columns = iris.select_dtypes(include=['float64', 'int64']).columns
sns.heatmap(iris[numeric_columns].corr(), annot=True)
plt.title('Heatmap')

plt.show()
```
