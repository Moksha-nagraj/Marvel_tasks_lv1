# Decision Trees
A Decision Tree is a supervised learning algorithm used for both regression and classification tasks. It's a predictive modeling tool that learns a hierarchical structure of decision rules from the data.


In this dataset ***salaries.csv*** we predict if person's salary more than 100 based on the company, his job title and degree.

```python
import pandas as pd
df=pd.read_csv("salaries.csv")
df.head()
```

```python
# Divide it between the target variable and independent variable datafrane inputs
inputs = df.drop('salary_more_then_100k',axis='columns')
target = df['salary_more_then_100k']
```

```python
inputs
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder
# Create objects of this class
le_company= LabelEncoder()
le_job= LabelEncoder()
le_degree = LabelEncoder()
```

```python
inputs['company_n']= le_company.fit_transform(inputs['company'])
inputs['job_n']= le_company.fit_transform(inputs['job'])
inputs['degree_n']= le_company.fit_transform(inputs['degree'])
inputs.head()
```

```python
inputs_n = inputs.drop(['company','job','degree'],axis='columns')
inputs_n
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(inputs_n,target)
```

```python
model.score(inputs_n,target)
```

```python
model.predict([[2,0,1]])
```
